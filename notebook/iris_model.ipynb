```python
y = dataset.iloc[:, 4].values
x = dataset.iloc[:, :4].values

#my test indices are first 10 and the last 20
test_indices = np.array(list(range(10)) + list(range(130,150)))
#set all of them to true
mask = np.ones(150,dtype=bool)
#except test_indices
mask[test_indices] = False

one_column = np.ones((150, 1))

x = np.concatenate((x, one_column), axis=1)

#inputs
X_train = x[mask, :]
X_test = x[~mask, :]

#expected output
y_train = y[mask]
y_test = y[~mask]
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X=X_train, y=y_train)
```

```python
log_reg.predict(X_test), y_test
```

```python
y_predictions = log_reg.predict_proba(X_test)
y_predictions
```

```python
from sklearn.metrics import log_loss

log_loss_score = log_loss(y_test, y_predictions)
log_loss_score
```
